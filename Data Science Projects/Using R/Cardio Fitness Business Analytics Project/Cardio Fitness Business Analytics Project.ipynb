#### EDA

```python
# Violin Plot for Education
fig, ax = plt.subplots()
sns.violinplot(x=working_data.Education, orient='h', ax=ax)
```

```python
# Violin Plot for Usage
fig, ax = plt.subplots()
sns.violinplot(x=working_data.Usage, orient='h', ax=ax)
```

```python
plt.figure(figsize=(15,7))
sns.heatmap(working_data.corr(), annot=True, cmap="rocket_r") 
```

```python
fig, ax = plt.subplots()
sns.boxplot(x=working_data.Income, y=working_data.Product, ax=ax).set(title='PRODUCT BY INCOME')
```

```python
fig, ax = plt.subplots()
sns.boxplot(x=working_data.Education, y=working_data.Product, ax=ax).set(title='PRODUCT BY EDUCATION')
```

```python
fig, ax = plt.subplots()
sns.boxplot(x=working_data.Miles, y=working_data.Product, ax=ax).set(title='PRODUCT BY MILES')
```

```python
fig, ax = plt.subplots()
sns.boxplot(x=working_data.Usage, y=working_data.Product, ax=ax).set(title='PRODUCT BY USAGE')
```

```python
# Pointplots
fig, ax = plt.subplots()
sns.pointplot(x=working_data.Education, y=working_data.Income, hue=working_data.Product).set(title='EDUCATION BY INCOME')
```

```python
#scatter plot between income, age ,product and usage
sns.relplot(x="Age", y="Income", hue="Product", size="Usage",
            sizes=(40, 400), alpha=.5, palette="nipy_spectral",
            height=6, data=working_data).set(title='INCOME BY AGE ,PRODUCT AND USAGE');
```
